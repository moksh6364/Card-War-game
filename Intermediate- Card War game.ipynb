```python
import random

suits = ("Clubs","Hearts","Spades","Diamonds")
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 
          'King':13, 'Ace':14}
```

```python
class Card():
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of "+ self.suit
```

```python
class Deck:
    
    def __init__(self):
        
        self.deck_all_card = []
        
        for suit in suits:
            for rank in ranks:
                new_card = Card(suit,rank)
                
                self.deck_all_card.append(new_card)
                
    def shuffle(self):
        random.shuffle(self.deck_all_card)
        
    def pick_one(self):   #once card pop-ed, it is removed from list
        return self.deck_all_card.pop()
```

```python
class Player:
    
    def __init__ (self, name):
        self.name = name
        self.deck_all_card = []
        
    def add_card(self,new_card):
        if type(new_card) == type([]):  # to add multiple cards
            self.deck_all_card.extend(new_card)
        else:                           # to add sinle card
            self.deck_all_card.append(new_card)
    
    def remove_card(self):
        return self.deck_all_card.pop()
    
    def __str__(self):
        return f'Player {self.name} has {len(self.deck_all_card)} cards.'
```

```python
player_one = Player("one")
player_two = Player("two")

deck = Deck()
deck.shuffle()

for i in range(26):
    player_one.add_card(deck.pick_one())
    player_two.add_card(deck.pick_one())


game = True
rounds = 0

while game:
    rounds += 1
    print( f"Round {rounds}")
    
    if len(player_one.deck_all_card) == 0:
        print("Player one deck is empty! Player Two wins.")
        game = False
        break
        
    if len(player_two.deck_all_card) == 0:
        print("Player two deck is empty! Player One wins.")
        game = False
        break
        
    player_one_deck = []
    player_one_deck.append(player_one.remove_card())
    
    player_two_deck = []
    player_two_deck.append(player_two.remove_card())
    
    war = True
    
    while war:
        
        if player_one_deck[-1].value > player_two_deck[-1].value:
            player_one.add_card(player_one_deck)
            player_one.add_card(player_two_deck)
            war = False
            
        elif player_one_deck[-1].value < player_two_deck[-1].value:
            player_two.add_card(player_one_deck)
            player_two.add_card(player_two_deck)
            war = False
            
        else: 
            print("WAR!")
            
            if len(player_one.deck_all_card) < 5:
                print("Player ONE unable to declare war! Player Two wins.")
                game = False
                break
                
            elif len(player_two.deck_all_card)< 5:
                print("Plater TWO unable to declare war! Player One wins.")
                game = False
                break
                
            else:
                for i in range(5):
                    player_one_deck.append(player_one.remove_card())
                    player_two_deck.append(player_two.remove_card())
```
